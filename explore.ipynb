# Labelling for evaluation

```python
import json
from pathlib import Path
import textwrap
from IPython.display import display, Markdown

CANDIDATES_PATH = Path("data/eval/candidates.jsonl")
QID = "q0010"  # 👈 change this each time

rows = [
    json.loads(l)
    for l in CANDIDATES_PATH.read_text(encoding="utf-8").splitlines()
    if l.strip()
]

row = next(r for r in rows if r["qid"] == QID)

display(Markdown(f"## {QID}"))
display(Markdown("### Query"))
display(Markdown(textwrap.fill(row["query"], width=110)))
display(Markdown("---"))
display(Markdown("### Candidates (in retrieval order)"))

for i, c in enumerate(row["candidates"], 1):
    display(Markdown(f"**{i}. rider_id={c['rider_id']} · score={c['score']:.4f}**"))
    summary = (c.get("summary") or "").replace("\n", " ").strip()
    display(Markdown(textwrap.fill(summary, width=110)))
```
